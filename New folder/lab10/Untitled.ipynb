```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt # used for plotting graph
plt.style.use("seaborn")
```

```python
a=0.6
```

```python
# step one take x and y data samples
xtrain=np.array(list(range(3,33)))
xtrain
```

```python
#increase for the dimension of xtrain 
xtrain=xtrain[:,np.newaxis]
xtrain
```

```python
len(xtrain)
```

```python
# intializing y train 
ytrain=np.array([1,2,1,1,1,4,5,6,7,8,9,10,12,11,11,13,14,16,15,17,18,19,21,22,23,24,25,26,27,28])
len(ytrain)
```

```python
plt.scatter(xtrain,ytrain)
```

```python
xtest=np.array([i/10 for i in range(400)]) # taking xo values 
xtest
```

```python
xtest=xtest[:,np.newaxis]
```

```python
ytest=[]# the object is find the ytest values and plot it on programs 
```

```python
for i in range(len(xtest)):
    wts=np.exp(-np.sum((xtrain-xtest[i])**2,axis=1)/2*(a)**2)
    w=np.diag(wts)
    format1=np.linalg.inv(xtrain.T.dot(w).dot(xtrain))
    predict=format1.dot(xtrain.T.dot(w).dot(ytrain))
    final=xtest[i].dot(predict)
    ytest.append(final)
ytest
```

```python
plt.scatter(xtrain,ytrain,color="green")
plt.scatter(xtest,ytest,color="red")
plt.show()
```
