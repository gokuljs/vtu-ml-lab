```python
import pandas as pd
import numpy as np
```

```python
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv("prog8-Kmeans_EM.csv")
df.head()
```

```python
f1=df['V1'].values
f2=df['V2'].values
x=np.array(list(zip(f1,f2)))
```

```python
x
```

```python
plt.scatter(f1,f2)
```

```python
k=3
kmean=KMeans(n_clusters=k)
kmean=kmean.fit(x)
labels=kmean.predict(x)
centroid=kmean.cluster_centers_
```

```python
centroid
```

```python
plt.scatter(x[:,0],x[:,1],c=labels,cmap="rainbow")
plt.scatter(kmean.cluster_centers_[:,0],kmean.cluster_centers_[:,1],marker="*",color="black")
```

```python
gnn=GaussianMixture(n_components=3)
gnn=gnn.fit(x)
labels=gnn.predict(x)
# probe=gnn.
plt.scatter(x[:,0],x[:,1],c=labels,s=50,cmap="rainbow")
```

```python
probe=gnn.probe
```
