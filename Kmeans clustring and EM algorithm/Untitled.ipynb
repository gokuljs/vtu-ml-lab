```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("prog8-Kmeans_EM.csv")
df.head()
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
len(df['V1'].values)
```

```python
f1=df['V1'].values
f2=df['V2'].values
x=np.array(list(zip(f1,f2)))
plt.scatter(f1,f2,c="black",s=7)
plt.title("datasets")
```

```python
# now using kmeans clustring 
k=3# selecting number of clusters
kmeans=KMeans(n_clusters=k)
kmeans=kmeans.fit(x)# fitting the dataset to jmeans cluster
labels=kmeans.predict(x)# predecting the labels 
centroids=kmeans.cluster_centers_# calculatinng the centroid 
```

```python
print(centroids)
```

```python
plt.title('Kmeans')
plt.scatter(x[:,0],x[:,1],c=labels,cmap="rainbow",s=9)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker="*",color="black")
```

```python
# for exmaple let us take clusters as 4
k=4
kmeans=KMeans(n_clusters=k)
kmeans=kmeans.fit(x)
labels=kmeans.predict(x)
centroid=kmeans.cluster_centers_
```
