```python
ytest=[]
a=0.5
```

```python
for i in range(len(xtest)):
    wts=np.exp(-np.sum((xtrain-xtest[i])**2,axis=1))/2*a**2
    w=np.diag(wts)
    form=np.linalg.inv(xtrain.T.dot(w).dot(xtrain))
    pred=form.dot(xtrain.T).dot(w).dot(ytrain)
    prediction=xtest[i].dot(pred)
    ytest.append(prediction)
```
